```python
import numpy as np
import matplotlib.image as img_pyp
import math
import matplotlib.pyplot as plt

#im = img.imread("Pictures/coins.png")

# im = Image.open()
def gamma_transform(c, f, gamma): 
    """ 
        c: constant 
        r: input image 

    """ 
    return c * np.array(f ** gamma, dtype = 'uint8') 

# c nh gc  
img = img_pyp.imread('../Image/Fig0309(a)(washed_out_aerial_image).tif') 

# lm trn nh gc float -> int  visualize 

img = np.array(img, dtype = np.uint8) 

# show nh gc
plt.imshow(img)
plt.title("Anh goc") 
plt.show() 

 

# Th 3 gi tr gamma 

# for gamma in [0.3, 3, 5]: 
# nh c bin i vi hm m 
gamma = 1
gamma_corrected = gamma_transform(2, img, gamma) 

    # visualize nh  
title = "Anh co gamma = 3" 
plt.imshow(gamma_corrected)
plt.title(title) 
plt.show() 
```

```python
import matplotlib.image as img
import numpy as np
import math
import matplotlib.pyplot as plt

# Read an image
image = img.imread('../Image/Fig0309(a)(washed_out_aerial_image).tif')
   
c = 255 / np.log(1 + np.max(image))
d= 0.0022 
exponential_image = c * (pow(1 + d, image) - 1)

exponential_image = np.array(exponential_image, dtype = np.uint8)

plt.imshow(image, cmap='gray')
plt.show()
plt.imshow(exponential_image, cmap='gray')
plt.show()
```
