```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
#negative transformation function
def neg_trans(img):
  #get width and height of image
  img = np.array(img)
  width,height = img.shape
  # get number of brightness levels of image
  L = 256
  for x in range(width):
    for y in range(height):
      I=img[x,y]
      #s=(L-1)-r
      S = (L - 1) - I 
      #replace the pixel
      img[x,y]  = S
  return img
```

```python
from PIL import Image 
image = Image.open("../Image/page.png")
```

```python
neg_image = neg_trans(image)
```
